```python
L_csv_path = os.path.join(PATH, L_FILE)
dataset_L = pd.read_csv(L_csv_path, dtype={"": int, "x": 'float64'})  

display(dataset_L.head())
display(dataset_L.info())
display(dataset_L.describe())
```

```python
import numpy as np

shuffled_indices = np.random.permutation(len(dataset))
test_set_size = int(len(dataset) * 0.20)
test_indices = shuffled_indices[:test_set_size]
train_indices = shuffled_indices[test_set_size:]
dataset.iloc[train_indices]
dataset.iloc[test_indices]

train_set = dataset.iloc[train_indices]
test_set = dataset.iloc[test_indices]

display(train_set.head())
display(test_set.head())
```

```python

from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(dataset, dataset_L["x"]):
    strat_train_set = dataset.loc[train_index]
    train_zero = strat_train_set.loc[dataset_L["x"] == 0]
    train_one = strat_train_set.loc[dataset_L["x"] == 1]
    
    strat_train_L_set = dataset_L.loc[train_index]
    train_L_zero = strat_train_L_set.loc[dataset_L["x"] == 0]
    train_L_one = strat_train_L_set.loc[dataset_L["x"] == 1]
    
    strat_test_set = dataset.loc[test_index]
    test_zero = strat_test_set.loc[dataset_L["x"] == 0]
    test_one =strat_test_set.loc[dataset_L["x"] == 1]
    
    strat_test_L_set = dataset_L.loc[test_index]
    test_L_zero = strat_test_L_set.loc[dataset_L["x"] == 0]
    test_L_one = strat_test_L_set.loc[dataset_L["x"] == 1]
    
display("done")
```
